```python
config = bittensor.dendrite.config()
config.dendrite.max_active_receptors = 2000
config.dendrite.max_worker_threads = 2000
config.dendrite.requires_grad = False
wallet = bittensor.wallet(name = 'const', hotkey = 'Nero')
dend = bittensor.dendrite(config = config, wallet = wallet)
```

```python
query_res = {}
```

```python
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor
bittensor.logging(debug=True)

text = 'cat dog'
def query_uid(uid ):
    if uid in query_res:
        return
    else:
        query_res[uid] = dend.forward_text( inputs = text, endpoints = meta.endpoints[uid])

results = []
with ThreadPoolExecutor(max_workers=10) as executor:
    results = list(tqdm(executor.map(query_uid, meta.uids.tolist()), total=len(meta.uids.tolist())))
    
```
